```python
from datetime import datetime, timedelta
import xarray as xr
dt = datetime(day=27,month=1,year=2021,hour=1)
# dt = datetime(day=28,month=5,year=2019,hour=23)
dt
```

```python
base_url = 'https://www.ncei.noaa.gov/thredds-ocean/dodsC/ghrsst/L3C/AMERICAS/GOES16/OSISAF/'

specific_url = dt.strftime('%Y')+'/'+dt.strftime('%j')+'/'+dt.strftime('%Y')+dt.strftime('%m%d')+\
    dt.strftime('%H%M%S')+'-OSISAF-L3C_GHRSST-SSTsubskin-GOES16-ssteqc_goes16_'+dt.strftime('%Y')+\
    dt.strftime('%m%d')+'_'+dt.strftime('%H%M%S')


url_end = '-v02.0-fv01.0.nc'
```

```python
url = base_url+specific_url+url_end
url
```

```python
ds_sst = xr.open_dataset(url).sel( lat=slice(29, 38), lon=slice(-82.5,-71))
```
